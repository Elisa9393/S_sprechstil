The index of the midlevel descriptors are extracted from cache.

```python
fex_mld = mld.MLD()
mld_df = fex_mld.extract_from_index(index=df_alexa, cache_path='./tmp/mld')
```

Calculate opensmile features by creating smile object, declaring features that should be extracted (here GeMAPSv01b) and declaring the level. 

```python
smile_lld = opensmile.Smile( #smile object
    feature_set = opensmile.FeatureSet.GeMAPSv01b,
    feature_level = opensmile.FeatureLevel.LowLevelDescriptors, #low level descriptors
)

smile_functionals = opensmile.Smile( #smile object
    feature_set = opensmile.FeatureSet.GeMAPSv01b,
    feature_level = opensmile.FeatureLevel.Functionals, #functionals
)
```

Return shape of the list of filenames: Analyzes functionals. Low level descriptors are multiplied with certain functionals

```python
funct_df = smile_functionals.process_files(df_alexa.index)
print(funct_df.shape)
```

Return shape of the list of filenames: Analyzes low level descriptors. There are less columns because not so many low level descriptors in gmaps version.

```python
lld_df = smile_lld.process_files(df_alexa.index)
print(lld_df.shape)
```

## Google Speech Recognition

Initalizing Google Speech Recognition in German.

```python
import asr as speechrec
asr = speechrec.ASR('deu')
df_asr = asr.extract_from_index(mld_df, tmp='./tmp',  margin=0.2)
```

Giving out the transcriptions of the first segments

```python
df_asr.head()
```

Accessing text_features.py to start the speech recognition. Use the try - except to treat errors.

```python
fex = tf.PsyText(language="deu")
ling_df = pd.DataFrame()
t_all = ''
for t in df_asr.asr:
    print(t)
    t_all = t_all+' '+t
    try:
        feat = fex.extract_from_string(t)
    except ZeroDivisionError:
        feat = pd.DataFrame()
    ling_df = ling_df.append(feat)
```
